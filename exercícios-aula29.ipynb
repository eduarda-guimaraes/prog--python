```python
import pandas as pd
```

 1. Carregar o dataset no Pandas

```python
df = pd.read_csv('/content/dados.csv')
print(df)
```

2. Exibir as primeiras linhas para inspecionar os dados

```python
primeiraLin = df.head(3)
print(primeiraLin)
```

3. Identificar valores ausentes e substituí-los de forma adequada


```python
# Idade
medianaIdade = df['Idade'].median()
df.loc[df['Idade'].isna(), 'Idade'] = medianaIdade
print('Idade:')
print(df['Idade'])
# Gasto Mensal
medianaGastoMensal = df['Gasto Mensal'].median()
df.loc[df['Gasto Mensal'].isna(), 'Gasto Mensal'] = medianaGastoMensal
print('\nGasto Mensal:')
print(df['Gasto Mensal'])
```

4. Corrigir tipos de dados (exemplo: converter "150,0" para float)

```python
# Idade
df['Idade'] = df['Idade'].astype(int)
print(df['Idade'])

# Gasto Mensal
df['Gasto Mensal'] = df['Gasto Mensal'].astype(str).str.replace(',','.').astype(float)
print(df['Gasto Mensal'])
```

5. Remover duplicatas

```python
df = df.drop_duplicates()
print(df)
```

6. Corrigir inconsistências nas categorias (ex: padronizar "Fem" para "F")


```python
from re import M
df['Gênero'] = df['Gênero'].replace({
    "Fem": "F",
    "Masc": "M"
})
print(df['Gênero'])
```

7. Descrever estatísticas básicas após os ajustes


```python
df.describe(include='all')
```

8. Gravar os resultados em um arquivo csv

```python
df.to_csv('novos_dados.csv', index=False)
```
